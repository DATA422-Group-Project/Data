```r
library(dplyr)
library(tidyverse)

dat = read.csv("owid-covid-data .csv")
dat$id = 1:nrow(dat)

cat("Latest date available:",max(dat$date)) #latest date available.
#Use a slightly earlier date so it's more likely data will be available for more countries.
date_ = "2021-09-03"

latest = filter(dat, dat$date == date_)

#Data frame of only 'total tests' data.
tests = data.frame(id = latest$id,total_tests = latest$total_tests)

#Remove missing values.
tests = na.omit(tests)

#How many countries have 'total tests' data for the chosen date.
cat("# of countries with total_tests data:",nrow(tests))
tests = left_join(tests,latest,by="id")
tests_df = data.frame(country=tests$location,total_tests=tests$total_tests.x,tests_per_1000=tests$total_tests_per_thousand)
head(tests_df)
```

Or could just get the latest date available for every country:

```r
dat$id = 1:nrow(dat)
test = data.frame(id = dat$id,tests = dat$total_tests)
test = na.omit(test)

test = left_join(test,dat,by="id")

latest = test %>%
  group_by(location) %>%
  summarize(date=max(date))

total_tests = left_join(latest,test,by=c("location","date")) %>%
  select(location,tests,total_tests_per_thousand,date) %>%
  as.data.frame()

total_tests
```
