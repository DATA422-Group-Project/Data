```r
library(tidyverse)
library(dplyr)
library(skimr)
library(stringr)
library(rvest)
```

```r
url_owid_survey <- "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv"

covid_survey <- read_csv(url_owid_survey)
```

```r
covid_df<-covid_survey %>% group_by(iso_code)%>%
    select(iso_code,location,total_vaccinations,people_vaccinated,people_fully_vaccinated,total_tests_per_thousand)
covid_df
```

```r
covid_vaccine<-covid_df %>% filter(!grepl("OWID_", iso_code))
covid_vaccine
```

```r
vaccine_data<-covid_vaccine %>%
    group_by(location, iso_code)%>%
    summarise_if(is.double, max)%>%
    mutate(half_vaccinated = (people_vaccinated - people_fully_vaccinated))%>%
    distinct() %>%
    slice()
```
