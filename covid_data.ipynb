```r
library(tidyverse)
library(dplyr)
library(skimr)
library(stringr)
library(rvest)
```

```r
url_cases <- "https://covid19.who.int/WHO-COVID-19-global-table-data.csv"

cases_df <- read_csv(url_cases)
```

```r
str(cases_df)
```

```r
cases_df
```

```r
cases_df<-cases_df %>% 
    select(Name, 'Cases - cumulative total', 'Deaths - cumulative total') %>%
    rename(country_name = Name, 
           total_cases = 'Cases - cumulative total',
           total_deaths = 'Deaths - cumulative total') %>%
    filter(!country_name=='Global')
cases_df
```

```r
library(countrycode)
```

```r
country_code <- countrycode(cases_df$country_name, 'country.name', 'iso3c')
```

```r
cases_df <- cases_df %>%
    add_column(country_code = if_else(.$country_name == countrycode(cases_df$country_name, 'country.name', 'iso3c'), country_code, country_code),
              .before="country_name")
```

```r
country = c("Bonaire", "Saba", "Sint Eustatius")
c <- cases_df %>% filter(country_name %in% country) %>%
     summarize_if(is.numeric, sum, na.rm=TRUE) %>%
     mutate(country_name = "Bonaire,Saba,Sint Eustatius")
```

```r
c<- c %>% select(country_name, total_cases, total_deaths)
```

```r
url_vaccines <- "https://covid19.who.int/who-data/vaccination-data.csv"

vaccines_df <- read_csv(url_vaccines)
```

```r
vaccines_df
```

```r
str(vaccines_df)
```

```r
vaccines_df <- vaccines_df %>% 
    select(ISO3, COUNTRY, TOTAL_VACCINATIONS,
           PERSONS_VACCINATED_1PLUS_DOSE,
           PERSONS_FULLY_VACCINATED, 
           TOTAL_VACCINATIONS_PER100) %>%
    rename(country_code = ISO3,
           country_name = COUNTRY,
           total_vaccinations = TOTAL_VACCINATIONS,
           vaccinated_1dose = PERSONS_VACCINATED_1PLUS_DOSE,
           fully_vacccinated = PERSONS_FULLY_VACCINATED,
           total_vaccinations_per100 = TOTAL_VACCINATIONS_PER100
           )
```

```r
vaccines_df[is.na(vaccines_df)] <- 0 
```

```r
vaccines_df
```

```r
covid_df <- left_join(cases_df, vaccines_df, by=c('country_code', 'country_name'))
```

```r
covid_df
```

```r
covid_df <- mutate_at(covid_df, c(3:8), ~replace(., is.na(.), 0))
```

```r
covid_df
```

```r
url_OxCGRT <- "https://raw.githubusercontent.com/OxCGRT/covid-policy-tracker/master/data/OxCGRT_latest.csv"

stringency_df <- read_csv(url_OxCGRT)
```

```r
head(stringency_df)
```

```r
stringency_df<- stringency_df %>% 
    select(CountryCode, StringencyIndex) %>%
    rename(country_code = CountryCode, 
           stringency_index = StringencyIndex) 
```

```r
options(warn=-1)
stringency_df <- stringency_df %>%
    group_by(country_code)%>%
    summarise_if(is.numeric, max, na.rm=T)%>%
    distinct() %>%
    slice()
options(warn=0)
```

```r
stringency_df
```

```r
covid_str_df <- left_join(covid_df, stringency_df, by="country_code")
```

```r
covid_str_df
```

```r
covid_str_df$stringency_index[is.na(covid_str_df$stringency_index)]<-0
```

```r
covid_str_df
```

```r
write.csv(covid_str_df, "Covid_Data.csv")
```
